```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-')
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],':')
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'-.',color = 'r')
plt.xlabel('xlabel',fontsize = 16)
plt.ylabel('ylabel',fontsize = 16)
```

```python
lin_num = np.arange(0,10,0.5)
```

```python
plt.plot(lin_num,lin_num,'r--')
```

```python
plt.plot(lin_num, lin_num ** 3, 'go')
```

```python
x = np.linspace(-10,10)
```

```python
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,linewidth = 3.0)
```

```python
plt.plot(x,y,color = 'r', linestyle =':' , marker  = 'o', markerfacecolor ='b', markersize = 3.0)
plt.title('lin label')
plt.text(0,0,'lin')
plt.grid(True)
plt.annotate('lin',xy=(-5,0),xytext=(-2,0.3),arrowprops = dict(facecolor='black',shrink=0.05,headlength= 20,headwidth = 20))
```

```python
line = plt.plot(x,y)
plt.setp(line , color = 'r', linewidth = 2.0, alpha = 0.9)
```

```python
plt.subplot(211)
plt.plot(x, y, color = 'r')

plt.subplot(212)
plt.plot(x, y, color = 'b')
```

```python
plt.subplot(121)
plt.plot(x, y, color = 'r')

plt.subplot(122)
plt.plot(x, y, color = 'b')
```
