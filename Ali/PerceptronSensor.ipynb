```python
class Perception:
    
    def __init__(self, alpha, times):
        
        self.alpha = alpha
        self.times = times
        
    def step(self, z):
        
        return np.where(z >= 0 ,1 , -1)
    
    def fit(self, X, y):
        
        X = np.asarray(X)
        y = np.asarray(y)
        
        self.paraWeight = np.zeros( 1 + X.shape[1])
        self.loss = []
        
        for i in range(self.times):
            
            localLoss = 0
            
            for x, target in zip(X, y):
                
                y_hat = self.step(np.dot(x, self.paraWeight[1:])) + self.paraWeight[0]
                localLoss += y_hat != target
                
                self.paraWeight[0] += self.alpha * (target - y_hat)
                self.paraWeight[1:] += self.alpha * (target - y_hat) * x
                
            self.loss.append(localLoss)
            
    def predict(self, X):
        
        return self.step(np.dot(X, self.paraWeight[1:]) + self.paraWeight[0])
```
