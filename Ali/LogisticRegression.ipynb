```python
class LogisticRegression:
    
    def __init__(self, alpha, times):
        
        self.alpha = alpha
        self.times = times
        
    def sigmoid(self, z):
        
        return 1.0/(1.0 + np.exp(-z))
    
    def fit(self, X, y):
        
        X = np.asarray(X)
        y = np.asarray(y)
        
        self.paraWeight = np.zeros(1 + X.shape[1])
        
        self.loss = []
        
        for in range(self.times):
            
            z = np.dot(X, self.paraWeight[1:]) + self.paraWeight[0]
            p = sigmoid(z)
            cost = - np.sum(y * np.log(p) + (1-y)*np.log(1-p))
            self.lost.append(cost)
            
    def predictProb(self, X):
        
        X = np.asarray(X)
        z = np.dot(X, self.paraWeight[1:] ) + self.paraWeight[0]
        p = self.sigmoid(z)
        p = p.reshape(-1,1)
        
        return np.concatenate([1-p, p], axis = 1)
    
    def predict(self, X):
        
        return np.argmax(self.predict_proba(X), axis = 1)
```
