```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
bostonData = pd.read_csv(r"boston.csv")
```

```python
bostonData
```

```python
class LinearRegression:
    
    def fit(self, X, y):
        
        X = np.asmatrix(X.copy())
        
        y = np.asmatrix(y).reshape(-1,1)
        
        self.para = (X.T * X).I *X.T *y
        
    def predict(self, X):
        
        X= np.asmatrix(X.copy())
        
        result = X * self.para
        
        return np.array(result).ravel()
```

```python
sampleData = bostonData.sample(len(bostonData), random_state = 0)
```

```python
sampleData
```

```python
train_X = sampleData.iloc[:400, :-1]
train_Y = sampleData.iloc[:400, -1]
test_X = sampleData.iloc[400:, :-1]
test_Y = sampleData.iloc[400:, -1]
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_X , train_Y)
```

```python
result=lr.predict(test_X)
```

```python
display(np.mean((result - test_Y) ** 2 ))
```

```python
biasColumn = sampleData.columns.insert(0, "bias")
sampleDataWithBias = sampleData.reindex(columns = biasColumn, fill_value= 1 )
```

```python
sampleDataWithBias
```

```python
trainWithBias_X = sampleDataWithBias.iloc[:400, :-1]
trainWithBias_Y = sampleDataWithBias.iloc[:400, -1]
testWithBias_X = sampleDataWithBias.iloc[400:, :-1]
testWithBias_Y = sampleDataWithBias.iloc[400:, -1]
```
