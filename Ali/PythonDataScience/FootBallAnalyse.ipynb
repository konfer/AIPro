```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams["font.family"] = "SimHei"
mpl.rcParams["axes.unicode_minus"] = False
```

```python
playerData = pd.read_csv(r"FullData.csv",index_col= "Name" ,parse_dates=["Club_Joining","Birth_Date"])
```

```python
pd.set_option("max_columns", 100)
```

```python
playerData.head(20)
```

```python
playerData.info()
```

```python
playerData.drop(playerData[playerData.loc[:,"Club_Kit"].isnull()].index, inplace = True)
```

```python
playerData.duplicated().any()
```

```python
playerData.loc[:,"Height"] = playerData.loc[:,"Height"].map(lambda x: int(x.replace("cm", " ")))
```

```python
playerData.loc[:,"Weight"] = playerData.loc[:,"Weight"].map(lambda x: int(x.replace("kg", " ")))
```

```python
playerData.sample(20)
```

```python
playerData.describe()
```

```python
playerData[["Height", "Weight", "Rating"]].plot(kind = "kde")
```

```python
playerData.loc[:, "Preffered_Foot"].value_counts()
```

```python
playerData.loc[:, "Preffered_Foot"].value_counts().plot(kind = "bar")
```

```python
clubGroup = playerData.groupby("Club")["Rating"].agg(["count", "sum", "mean"])
```

```python
clubGroup = clubGroup[clubGroup["count"] > 20 ]
```

```python
clubGroup.sort_values("mean", ascending = False)
```

```python
countryGroup = playerData.groupby("Nationality")["Rating"].agg(["count", "sum", "mean"])
```

```python
countryGroup = countryGroup[countryGroup["count"] > 15 ]
```

```python
countryGroup.sort_values("mean", ascending = False)
```
